```python
# impor library
import pandas as pd
import folium
```

```python
# impor dataset
df = pd.read_csv('dataset.csv')
```

```python
# cek dataframe
df.head()
```

```python
# drop kolom yang tidak digunakan
df = df.drop(['ADDR_PCT_CD', 'CRM_ATPT_CPTD_CD', 'HADEVELOPT', 'HOUSING_PSA', 'JURISDICTION_CODE', 'KY_CD', 'LAW_CAT_CD', 'LOC_OF_OCCUR_DESC', 'OFNS_DESC', 'PARKS_NM', 'PD_CD', 'PD_DESC', 'PREM_TYP_DESC', 'STATION_NAME','SUSP_AGE_GROUP','SUSP_RACE','SUSP_SEX','TRANSIT_DISTRICT','VIC_AGE_GROUP','VIC_RACE','VIC_SEX','X_COORD_CD', 'Y_COORD_CD', 'Lat_Lon', 'CMPLNT_TO_DT', 'CMPLNT_TO_TM', 'RPT_DT'], axis=1)
```

```python
df.head()
```

```python
# menghapus row yang memiliki nilai Latitude dan Longitude null (NaN)
df = df[df['Latitude'].notnull() & df['Longitude'].notnull()]
```

```python
# memastikan Latitude dan Longitude tidak null
df.isnull().sum()
```

```python
# menggabungkan date dan time
df['CMPLNT_FR_DTTM'] = df[['CMPLNT_FR_DT', 'CMPLNT_FR_TM']].apply(lambda x: ' '.join(x), axis=1)
```

```python
df.head()
```

```python
df['CMPLNT_FR_DTTM'] = pd.to_datetime(df['CMPLNT_FR_DTTM'], format='%m/%d/%Y %H:%M:%S', errors='coerce')
```

```python
# drop kolom yang tidak digunakan
df = df.drop(['CMPLNT_FR_DT', 'CMPLNT_FR_TM'], axis=1)
```

```python
df.head()
```

```python
df['MONTH'] = df['CMPLNT_FR_DTTM'].apply(lambda x: x.month).fillna(0).astype(int)
```

```python
df.head()
```

```python
df.MONTH.value_counts()
```

```python
# menghapus row yang month == 0 (NaN)
df = df[df['MONTH'] != 0]
```

```python
df.head()
```

```python
# menggunakan data yang terjadi pada bulan februari saja
df = df[df['MONTH'] == 2]
```

```python
df.head()
```

```python
df.columns = ['id','boro','juris', 'patrol','latitude','longitude', 'datetime', 'month']
```

```python
df.head()
```

```python
df.to_csv(r'clean.csv', index=False)
```
